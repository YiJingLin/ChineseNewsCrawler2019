```python
import requests, time, random
from bs4 import BeautifulSoup
from tqdm import tqdm
```

# 1. crawl news link

```python
links = []

# https://www.epochtimes.com.tw/最新文章/201/中國/p/7
for i, cls in enumerate(['202/國際','201/中國']):
    for p in range(1, 21): # 1 ~ 20
        page = requests.get("https://www.epochtimes.com.tw/最新文章/%s/p/%d" % (cls, p))
        soup = BeautifulSoup(page.content, "lxml")
        tags = soup.find_all('div', {"class":"item_inner"})
        links += ['https://www.epochtimes.com.tw'+ tag.find_all('a')[-1]['href'] for tag in tags]
        links
        time.sleep(1.5+random.random())
        print(i*20+p, '/40')
```

```python
links = list(set(links))
print(len(links))
links[:10]
```

# 2. crawl news

```python
titles = []
dates = []
contents = []

for idx, link in enumerate(links):
    time.sleep(random.random())
    try:
        page = requests.get(link)
        soup = BeautifulSoup(page.content, "lxml")
        [s.extract() for s in soup('script')] # remove script

        title = soup.find_all('h1', {"class":"entry-title"})[0].text
        date = soup.find_all('time', {"class":"entry-date updated td-module-date"})[0]['datetime']
        content = soup.find_all('div', {"class":"td-paragraph-padding-1"})[0].text.replace('\n','')
        
        titles.append(title)
        dates.append(date)
        contents.append(content)
        print('[INFO] process success, (%d/600) link : %s' % (idx, link))
    except :
        print('[ERROR] process fail, (%d/600) link : %s' % (idx, link))
        
```

# 3. output csv

```python
import pandas as pd

if len(titles) == len(contents) and len(contents) == len(dates):
    df = pd.DataFrame({'title':titles, 'content':contents, 'date':dates})
    df = df[['date', 'title', 'content']]
    df.to_excel('epochtimes.xlsx', index=False)
    print('[INFO] the result has been saved successfully : ./epochtimes.xlsx')
else:
    print('[ERROR] mismatch lengths of titles, contents, and dates list')
```
